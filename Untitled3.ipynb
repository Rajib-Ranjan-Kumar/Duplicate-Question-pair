```python
df=df.dropna()
```

```python
df['q1_len'] = df['question1'].str.len()
df['q2_len']=df['question2'].str.len()
```

```python
df['q1_word'] = df['question1'].str.split().str.len()
df['q2_word'] = df['question2'].str.split().str.len()
```

```python
def common_w(text1, text2):
    s1 = set(text1.split())
    s2 = set(text2.split())
    return len(s1 & s2)

df['common_word'] = df.apply(
    lambda x: common_w(x['question1'], x['question2']),
    axis=1
)
```

```python
df['total_word']=df['question1'].str.split().str.len()+df['question2'].str.split().str.len()
```

```python
df['word_share']=df['common_word']/df['total_word']
```

```python
def jaccard_similarity(q1, q2):
    if not isinstance(q1, str) or not isinstance(q2, str):
        return 0
    s1, s2 = set(q1.split()), set(q2.split())
    return len(s1 & s2) / len(s1 | s2) if len(s1 | s2) != 0 else 0

df['jaccard_sim'] = [
    jaccard_similarity(q1, q2)
    for q1, q2 in zip(df['question1'], df['question2'])
]
```

```python
df['q1_len'] = df['question1'].fillna("").str.len()
df['q2_len'] = df['question2'].fillna("").str.len()
df['len_diff'] = abs(df['q1_len'] - df['q2_len'])
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def common_stopword_ratio(q1, q2):
    s1 = set(q1.split()) & stop_words
    s2 = set(q2.split()) & stop_words
    return len(s1 & s2) / (len(s1 | s2) + 1)

df['common_stopword_ratio'] = [
    common_stopword_ratio(q1, q2)
    for q1, q2 in zip(df['question1'], df['question2'])
]
```

```python
def first_word_same(q1, q2):
    try:
        return int(q1.split()[0] == q2.split()[0])
    except:
        return 0

df['first_word_same'] = [
    first_word_same(q1, q2)
    for q1, q2 in zip(df['question1'], df['question2'])
]
```

```python
q1 = df['question1']
q2 = df['question2']
```

```python
df['unique_word_diff'] = [
    abs(len(set(a.split())) - len(set(b.split())))
    for a, b in zip(q1, q2)
]
```

```python
def overlap_coeff(a, b):
    s1, s2 = set(a.split()), set(b.split())
    return len(s1 & s2) / (min(len(s1), len(s2)) + 1)

df['overlap_coeff'] = [overlap_coeff(a, b) for a, b in zip(q1, q2)]
```

```python
df['exact_match'] = [
    int(a.strip().lower() == b.strip().lower())
    for a, b in zip(q1, q2)
]
```

```python
def char_overlap(a, b):
    s1, s2 = set(a), set(b)
    return len(s1 & s2) / (len(s1 | s2) + 1)

df['char_overlap'] = [char_overlap(a, b) for a, b in zip(q1, q2)]
```

```python
def avg_word_len_diff(a, b):
    aw = sum(len(w) for w in a.split()) / (len(a.split()) + 1)
    bw = sum(len(w) for w in b.split()) / (len(b.split()) + 1)
    return abs(aw - bw)

df['avg_word_len_diff'] = [avg_word_len_diff(a, b) for a, b in zip(q1, q2)]
```

```python
df['word_count_ratio'] = [
    len(a.split()) / (len(b.split()) + 1)
    for a, b in zip(q1, q2)
]
```

```python
import string

def punct_diff(a, b):
    return abs(
        sum(c in string.punctuation for c in a) -
        sum(c in string.punctuation for c in b)
    )

df['punct_diff'] = [punct_diff(a, b) for a, b in zip(q1, q2)]
```

```python
df['first_word_match'] = [
    int(a.split()[0] == b.split()[0]) if a and b else 0
    for a, b in zip(q1, q2)
]

df['last_word_match'] = [
    int(a.split()[-1] == b.split()[-1]) if a and b else 0
    for a, b in zip(q1, q2)
]
```

```python
from nltk.util import ngrams

def bigram_overlap(a, b):
    bg1 = set(ngrams(a.split(), 2))
    bg2 = set(ngrams(b.split(), 2))
    return len(bg1 & bg2)

df['bigram_overlap'] = [bigram_overlap(a, b) for a, b in zip(q1, q2)]
```

```python
def trigram_overlap(a, b):
    tg1 = set(ngrams(a.split(), 3))
    tg2 = set(ngrams(b.split(), 3))
    return len(tg1 & tg2)

df['trigram_overlap'] = [trigram_overlap(a, b) for a, b in zip(q1, q2)]
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def stopword_diff(a, b):
    return abs(
        sum(w in stop_words for w in a.split()) -
        sum(w in stop_words for w in b.split())
    )

df['stopword_diff'] = [stopword_diff(a, b) for a, b in zip(q1, q2)]
```

```python
def content_overlap(a, b):
    s1 = set(w for w in a.split() if w not in stop_words)
    s2 = set(w for w in b.split() if w not in stop_words)
    return len(s1 & s2)

df['content_overlap'] = [content_overlap(a, b) for a, b in zip(q1, q2)]
```

```python
import difflib

df['sequence_ratio'] = [
    difflib.SequenceMatcher(None, a, b).ratio()
    for a, b in zip(q1, q2)
]
```

```python
def prefix_match(a, b):
    i = 0
    while i < min(len(a), len(b)) and a[i] == b[i]:
        i += 1
    return i

df['prefix_match_len'] = [prefix_match(a, b) for a, b in zip(q1, q2)]
```

```python
def suffix_match(a, b):
    i = 0
    while i < min(len(a), len(b)) and a[-1-i] == b[-1-i]:
        i += 1
    return i

df['suffix_match_len'] = [suffix_match(a, b) for a, b in zip(q1, q2)]
```

```python
def digit_overlap(a, b):
    return len(set(filter(str.isdigit, a)) & set(filter(str.isdigit, b)))

df['digit_overlap'] = [digit_overlap(a, b) for a, b in zip(q1, q2)]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(max_features=5000)

tfidf_matrix = tfidf.fit_transform(
    list(df['question1']) + list(df['question2'])
)

n = len(df)
q1_tfidf = tfidf_matrix[:n]
q2_tfidf = tfidf_matrix[n:]

df['tfidf_cosine'] = [
    cosine_similarity(q1_tfidf[i], q2_tfidf[i])[0][0]
    for i in range(n)
]
```
